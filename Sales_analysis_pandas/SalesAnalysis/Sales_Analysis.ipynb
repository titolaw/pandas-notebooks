### Sales Analysis

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn'
```

#### Merging 12 months os sales data into a single file

```python
# option 1:
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    temp = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, temp])
    
all_months_data.to_csv("all_data.csv", index=False)

df = pd.read_csv("all_data.csv")
df
```

```python
# option 2
frames = [pd.read_csv('./Sales_Data/Sales_January_2019.csv'), 
          pd.read_csv('./Sales_Data/Sales_February_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_March_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_April_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_May_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_June_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_July_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_August_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_September_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_October_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_November_2019.csv'),
          pd.read_csv('./Sales_Data/Sales_December_2019.csv')]

df = pd.concat(frames,ignore_index=True)
df
```

#### Data cleaning

```python
# drop NaN rows
nan_df = df[df.isna().any(axis=1)]
nan_df.head()
df = df.dropna(how = 'all')
df
```

#### Find 'Or' and delete it

```python
df = df[df['Order Date'].str[0:2] != 'Or']
df
```

#### Convert columns to the correct type

```python
# option 1
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int64')
df['Price Each'] = df['Price Each'].astype('float64')
```

```python
# option 2
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df.dtypes
```

#### Augment data with additional columns

#### Task 2: Add month column

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int32')
```

#### Task 3: Add sales column 

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

#### Task 4: Add a city column

```python
# Using .apply()
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

df.head()
```

#### Q1: what was the best month for sales? How much was earned that month?

```python
results = df.groupby('Month').sum()
results
```

```python
months = range(1,13)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD (US$)')
plt.xlabel('Month number')
plt.show()
```

#### Q2: What city had the highest number of sales?

```python
results = df.groupby('City').sum()
results
```

```python
cities = [city for city, df in df.groupby('City')]

plt.bar(cities,results['Sales'])
plt.xticks(cities, rotation = 'vertical',size=8)
plt.ylabel('Sales in USD (US$)')
plt.xlabel('City name')
plt.show()
```

#### Q4: What products are most often sold together?

```python
sub_df = df[df['Order ID'].duplicated(keep=False)]

sub_df['Grouped'] = sub_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

sub_df = sub_df[['Order ID', 'Grouped']].drop_duplicates()
sub_df.head()
```

```python
# reference: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in sub_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key,value)
```

#### Q5: What product sold the most? Why it sold the most?

```python
prod_group = df.groupby('Product')
quantity_ordered = prod_group.sum()['Quantity Ordered']

products = [product for product, df in prod_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity ordered')
plt.xlabel('Product')
plt.xticks(products, rotation='vertical',size=8)
plt.show()

```

```python
# Referece: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib
prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
