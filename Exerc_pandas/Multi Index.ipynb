### Extract index level values with the get_level_values method

```python
# criando multiIndex no read_csv
mc = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
mc.sort_index(inplace = True)
mc.head(3)
```

```python
# retornar valores de certo level de index
mc.index.get_level_values("Date")
mc.index.get_level_values(0)
```

```python
mc.index.get_level_values("Country")
mc.index.get_level_values(1)
```

### Change index level name with the set_names method

```python
mc = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
mc.sort_index(inplace = True)
mc.head(3)
```

```python
mc.index.set_names(names = ["Day","Location"], inplace = True)
mc.head(3)
```

```python
# mc.index.set_names(names = "Date", level = 0)
mc.index.set_names(names = "Date", level = "Day", inplace = True)
```

### The sort_index method on a multiIndex dataframe

```python
mc = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
mc.head(3)
```

```python
# different sorting at different indexes
mc.sort_index()
mc.sort_index(ascending = True)
mc.sort_index(ascending = False)
mc.sort_index(ascending = [True, False])
```

```python
# sorting um index e deixando o outro na ordem original
mc.sort_index(level = 0)
```

### Extract rows from a multiIndex dataframe

```python
mc = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
mc.head(3)
```

```python
# .loc eh ambiguo quanto a selecao de valores ou colunas no segundo argumento
mc.loc["2010-01-01", "Argentina"]
mc.loc["2010-01-01", "Price in US Dollars"]
```

```python
# metodo sugerido na documentacao do pandas, usar uma tuple, primeiro argumento rows e segundo columns
mc.loc[("2010-01-01", "Argentina"),"Price in US Dollars"]
mc.loc[("2010-01-01",)]
```

### The transpose method

```python
mc = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
mc.sort_index(inplace = True)
mc.head(3)
```

```python
mc = mc.transpose()
mc
```

```python
# dar preferencia a sintaxe de tuple(virgula no final se so um argumento)
# mc.loc[("Price in US Dollars",)]
# mc.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka")]
mc.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka"):("2010-01-01","Ukraine")]
```

### The .stack() method

```python
ws = pd.read_csv("worldstats.csv", index_col = ["country","year"])
ws
```

```python
# .stack takes the column base index, and moves to the row base index
ws.stack()
```

```python
# convertendo para dataframe
ws.stack().to_frame()
```

### The .unstack() method, part 1

```python
ws = pd.read_csv("worldstats.csv", index_col = ["country","year"])
ws.head(3)
```

```python
# transforma a coluna em index
s = ws.stack()
s.head(3)
```

```python
# pega o index mais profundo do dataset e o transforma em colunas
s.unstack()
s.unstack().unstack().unstack()
```

### The .unstack() method, part 2

```python
ws = pd.read_csv("worldstats.csv", index_col = ["country","year"])
ws.head(3)
```

```python
s = ws.stack()
s.head(3)
```

```python
# unstack aceita o index do index que se deseja transformar em colunas
s.unstack(0)
#s.unstack("year")
```

### The .unstack() method, part 3

```python
ws = pd.read_csv("worldstats.csv", index_col = ["country","year"])
s = ws.stack()
s.head(3)
```

```python
# unstack senquencial 
s.unstack(level = [0, 1])
```

### The pivot method

```python
sm = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sm["Salesman"] = sm["Salesman"].astype("category")
sm
```

```python
len(sm)
```

```python
sm.info()
```

```python
sm.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

### The pivot_table() method

```python
foods = pd.read_csv("foods.csv")
foods
```

```python
foods.pivot_table(values = "Spend", index = "Item", aggfunc = "mean")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"],columns = "City", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"],columns = ["Frequency","City"], aggfunc = "sum")
```

```python
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```
