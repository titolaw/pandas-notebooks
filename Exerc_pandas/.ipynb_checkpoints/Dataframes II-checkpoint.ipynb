### The .duplicated() method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
# marca a primeira ocorrencia como False, e o resto como True
# seleciona o primeiro valor
# df[df["First Name"].duplicated() == False]
# retorna todos os valores que tem duplicata
df[df["First Name"].duplicated(keep = False)]
```

```python
# selecionando todos os que nao sao duplicados (~ => inverte)
df[~df["First Name"].duplicated(keep = False)]
```

### The .drop_duplicates() method

```python
# funciona no dataframe em vez de serie, utilizando menos linhas de codigo
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df
```

```python
len(df)
```

```python
# exclui toda linha que for duplicada, keep mantem primeira ou ultima
df.drop_duplicates(subset = ["First Name"], keep = "first")
```

```python
# aceita multiplas entradas, remove as linhas que forem duplicadas em nome e time ao mesmo tempo:
df.drop_duplicates(subset = ["First Name","Team"])
```
