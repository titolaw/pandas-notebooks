```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

### Use the plot method to render a line chart

```python
bb = data.DataReader(name = "BB", data_source = "yahoo", start = "2007-01-01", end = "2020-12-31")
bb.head()
```

```python
bb.plot()
bb.plot(y = "High")
```

```python
bb["High"].plot()
```

```python
bb[["High", "Close"]].plot()
```

### Modifying plot aesthetics with matplotlib templates

```python
plt.style.available
```

```python
plt.style.use("ggplot")
bb.plot(y = "Close")
```

### Creating bar graphs to show counts

```python
bb.head(3)
```

```python
def rank_perfomance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_perfomance).value_counts().plot(kind = "barh")
```

### Creating a pie charts to represent proportions

```python
bb.head(3)
```

```python
average = bb["Close"].mean()
```

```python
def rank_perfomance(stock_price):
    if stock_price >= average:
        return "Above average"
    else:
        return "Below average"
```

```python
plt.style.use("dark_background")
bb["Close"].apply(rank_perfomance).value_counts().plot(kind = "pie", legend = True)
```
