### Drop rows with null values

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# removendo qualquer linha que tenha NaN
nba.dropna()
```

```python
# removendo linhas que seja so NaN
nba.dropna(how = "all")
```

```python
# removendo linhas com NaN em coluna especifica
nba.dropna(subset = ["College"])
nba.dropna(subset = ["College", "Salary"])
```

### Fill in missing dataframe values with the fillna method

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# nba.fillna(0)
# nba["College"].fillna("Unknown", inplace = True)
nba["College"] = nba["College"].fillna("Unknown")
```

```python
nba.tail()
```

```python
# nba["Salary"].fillna(value = 0, inplace = True)
nba["Salary"] = nba["Salary"].fillna(value = 0)
```

```python
nba.tail()
```

### The astype method I

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.tail()
```

```python
nba["Age"].hasnans
nba["Age"].astype("int")
nba["Age"] = nba["Age"].astype(int)
```

```python
nba.tail()
```

```python
# o metodo nao funciona se houve NaN values, deve-se substituir por outro valor antes
nba["Salary"].hasnans
nba["Salary"].fillna(value = 0)
nba["Salary"].astype(int).head()
```

### The astype method II

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.tail()
```

```python
# category column, coluna com poucos valores possiveis
# tranformar uma column em category columns diminui o consumo de memoria de um dataframe
```

```python
nba.dtypes
```

```python
len(nba)
nba.info()
```

```python
nba["Position"].nunique()
```

```python
nba["Position"].astype("category")
```

```python
# categoria armazena uma referenca a cada valor, em vez de varios valores, economizando espaco
nba["Position"] = nba["Position"].astype("category")
```

```python
nba.info()
```

### Rank values with the rank method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.head()
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba.head()
```

```python
nba["Salary"].rank()
nba["Salary"].rank(ascending = True)
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype(int)
nba.head()
```

```python
nba.sort_values("Salary", ascending = False)
```
