### Retrieve a group with the .get_group() method

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby("Sector")
ft.head(3)
```

```python
# extrai todos os valores de determinado grupo
sectors.get_group("Energy")
```

### Methods on the groupby object and dataframe columns

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby("Sector")
ft.head(3)
```

```python
# extrai todos os valores maximos (alfabeticamente) de cada grupo, arranjados pelo valor mais a esquerda, nesse caso Company
sectors.max()
```

```python
# retorna a soma de todos os valores numericos para cada setor
sectors.sum()
```

```python
sectors["Revenue"].sum()
sectors["Profits"].max()
sectors["Profits"].min()

sectors[["Profits", "Revenue"]].sum()
```

### Grouping by multiple columns

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby(["Sector", "Industry"])
ft.head(3)
```

```python
sectors.size()
sectors["Revenue"].sum()
sectors.sum()
```

### The .agg() method

```python
ft = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = ft.groupby(["Sector", "Industry"])
ft.head(3)
```

```python
# permite diferentes operacoes em cada coluna
sectors.agg({"Revenue" : "sum", 
              "Profits" : "sum",
             "Employees" : "mean"})
```

```python
sectors.agg(["size", "sum", "mean"])
```
